Exploratory Data Analysis


```python
print (df.columns)
```

On the basis of GENDER

```python
ax = sns.countplot(x = "Gender", data = df)
plt.title("Gender Comparison")
for bar in ax.containers:
    ax.bar_label(bar)
 
```

```python
sales_by_gender = df.groupby(["Gender"], as_index=False)["Amount"].sum()
print (sales_by_gender)
```

```python
ax = sns.barplot(x = "Gender", y = "Amount" , data = sales_by_gender)
for bar in ax.containers:
    ax.bar_label(bar)
```

From the above analysis we can say that Female has higher purchase power than female and has done more shopping in Diwali


```python
ax = sns.countplot(x = "Age Group" , data = df , hue = "Gender")
plt.title("Age Comparison")
for bar in ax.containers:
    ax.bar_label(bar)
plt.show()
```

```python
sales_by_age = df.groupby(["Age Group"] , as_index= False)["Amount"].sum().sort_values(by = "Amount",ascending=False)
print (sales_by_age)
```

```python
ax = sns.barplot(x = "Age Group", y = "Amount" , data = sales_by_age)
```

From the above analysis we can say that the highest numbers of buyers are woman and from the age group 26-35

```python
print (df.columns)
```

```python
ax = sns.countplot(x = "State", data = df)
sns.set(rc={"figure.figsize" : (20,5)})
for bar in ax.containers:
    ax.bar_label(bar)
```

```python
sales_by_state = df.groupby(["State"], as_index=False)['Orders'].sum().sort_values(by ="Orders", ascending=False).head(10)
ax = sns.barplot(x = "State", y ="Orders" , data = sales_by_state )
for bar in ax.containers:
    ax.bar_label(bar)
```

```python
sales_by_state = df.groupby(["State"], as_index=False)['Amount'].sum().sort_values(by ="Amount", ascending=False).head(10)
ax = sns.barplot(x = "State", y ="Amount" , data = sales_by_state )
for bar in ax.containers:
    ax.bar_label(bar)
```

From the above anlaysis we can say that the maximum orders and revenue are from uttar pradesh and lowest ordes and revenue are from Gujrat from the top 10 states
